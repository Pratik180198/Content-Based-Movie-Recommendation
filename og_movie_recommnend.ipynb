```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
def get_actor(obj):
    L=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter = counter + 1
        else:
            break
    return L
```

```python
movies.cast=movies.cast.apply(get_actor)
```

```python
movies.head()
```

```python
movies.crew[0]
```

```python
def get_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies.crew=movies.crew.apply(get_director)
```

```python
def stem(obj):
    z=[]
    for i in obj.split():
        z.append(stemming.stem(i))
    return " ".join(z)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
cv=TfidfVectorizer(max_features=10000,ngram_range=(1,3),stop_words='english')
```

```python
vectors = cv.fit_transform(movies['tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names()
```

```python
similarity=cosine_similarity(vectors)
```

```python
def recommend(movie_name):
    index = movies[movies['original_title'] == movie_name].index[0]
    distances = similarity[index]
    all_movies = sorted(list(enumerate(distances)),reverse=True,key=lambda x : x[1])[1:11]
    for i in all_movies:
        print(movies['original_title'].iloc[i[0]])
```

```python
recommend('Spider-Man')
```
